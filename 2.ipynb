```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 读取CSV数据集
data = pd.read_csv('/content/drive/MyDrive/dataset.csv')

# 可选：查看数据集的前几行，确保数据加载正确
print(data.head())

# 创建动作之间的关联性矩阵
action_matrix = data.pivot_table(index='user', columns='action', values='step', aggfunc='count', fill_value=0)

# 计算相关性矩阵
correlation_matrix = action_matrix.corr()

# 设置图像大小
plt.figure(figsize=(20, 20))
sns.set(font_scale=1.0)  # 调整字体大小

# 绘制热度图
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="YlGnBu")

# 添加标题
plt.title('Heat map of associativity between actions')

# 显示图像
plt.show()
```

```python
plt.savefig('/content/drive/MyDrive/heatmap.png', dpi=600, bbox_inches='tight')
```
