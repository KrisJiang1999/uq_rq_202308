```python
import pandas as pd
```

```python
# 读取CSV文件
file_path = "/content/drive/MyDrive/dataset.csv"  # 替换为您的文件路径
df = pd.read_csv(file_path)
```

```python
# 删除"timestamp"和"step"列
columns_to_remove = ["timestamp", "step"]
df = df.drop(columns=columns_to_remove)
```

```python
# 删除相同用户下连续重复的动作
df_cleaned = df[~((df["action"] == df["action"].shift()) & (df["user"] == df["user"].shift()))]
```

```python
# 删除连续重复行
df = df[df['user'] != df['user'].shift()]

# 提取包含'tab_other'的行索引
tab_other_indices = df[df['action'] == 'tab_other'].index

# 提取前三行和后三行的索引
selected_indices = set()
for index in tab_other_indices:
    selected_indices.update(range(index - 3, index + 3))

# 筛选出选定的行
df = df[df.index.isin(selected_indices)]
```

```python
# 保存处理后的数据到新文件
df_cleaned.to_csv('/content/drive/MyDrive/UQ/dataset.csv', index=False)
```
